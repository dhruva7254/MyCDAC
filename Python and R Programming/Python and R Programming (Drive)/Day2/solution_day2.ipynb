# Range function

range(start, end , step) #--> creates a range of integer values
using start, end and step
start : starting value of the range to be generated
default value of start = 0
end : ending value of range. Range will end at end-1 value
step : increment / decrement value for each element
default value of step = 1

```python
print(range(5)) # ONLY end value is given, start=0, step=1
```

```python
for idx in range(5): #start=0, step=1
    print(idx)
```

# Pattern Printing

### Diamond of stars

```python
n=5
sp=n//2
for i in range(1,n+1,2):
    print(" "*sp,'*'*i, sep="")
    sp = sp -1
```

### 10 pattern

```python
n=5
sp=0
for idx in range(n-1,-1,-1):
    print(" "*sp,'10'*idx,'1',sep="")
    sp = sp+1
```

10.	Write a program in python to find the sum of the series [ x - x^3 + x^5 –x^7+x^9-x^11+ ......].

Test Data :
Input the value of x :2 Input number of terms : 5

Expected Output :

The values of the series:

2

-8

32

-128

512

The sum = 410

```python
x=2
n=5
idx=1
total=0
for counter in range(1,n+1):
    series_elem = x ** idx
    if counter%2 == 0:
        #series_elem *= -1
        series_elem = -series_elem
    # print(series_elem)
    total += series_elem
    idx +=2
print(total)
```
