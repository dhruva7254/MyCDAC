# recurssive function

```python
count=3
def hello_fun():
    global count
    if count > 0: # breaking condition
        print("hello")
        count -=1
        hello_fun() # call the function itself
    else:
        pass # dont call the recurssive function!!!
hello_fun()
```

```python
def hello_fun(count):
    if count > 0: # breaking condition
        print("hello")
        hello_fun(count-1) # call the function itself
    else:
        pass # dont call the recurssive function!!!
hello_fun(3)
```

# function mutable parameter working

```python
def fun2(l:list):
    l[0]=10

l1=[3,4,5]
print(id(l1),l1)
fun2(l1)
print(id(l1), l1)
```

```python
def fun2(l:list):
    # l[0]=10
    l=[12,34]
    print("l:",id(l),l)

l1=[3,4,5]
print("before call l1:",id(l1),l1)
fun2(l1)
print("after call l1:",id(l1), l1)
```
