# Inheritance

## Inheriting attributes and methods

## Single class inheritance

```python
class A:
    def __init__(self):
        self.a=90
    def display(self):
        print("This is class A a=", self.a)
    def increment(self):
        self.a += 100
        print("Incremented value is ", self.a)

class B:
    def __init__(self):
        self.b="abcd"
```

```python
class C(A):
    def __init__(self):
        #call the __init__ of the base class
        # super().__init__()
        A.__init__(self)
        self.c=[9,8,7]
    def display_info(self):
        print(self.c, self.a)
    #over ride increment method to add extra functionality!
    def increment(self):
        self.a += 500
        print("In class C, increment by 500. New val is ", self.a)
```

```python
c1 = C()
c1.display_info()
```

```python
c1.display()
```

```python
c1.increment()
c1.display_info()
```

## Multiple inheritance

```python
class A:
    def __init__(self):
        print("In init of A")
        self.a=90
    def display(self):
        print("This is class A a=", self.a)
    def increment(self):
        self.a += 100
        print("Incremented value is ", self.a)

class B:
    def __init__(self):
        print("In init of B")
        self.b="abcd"
    def new_display(self):
        print("In class B. b=", self.b)
```

```python
class D(A,B):
    def __init__(self):
        print("In init of D")
        #call the __init__ of the base class
        # super().__init__()
        A.__init__(self)
        B.__init__(self)
        self.d={'k1':'v1'}
    def display_info(self):
        print(self.d, self.a)
    #over ride increment method to add extra functionality!
    def increment(self):
        self.a += 500
        print("In class D, increment by 500. New val is ", self.a)
```

```python
d1 = D()
```

```python
print(d1.a, d1.b, d1.d)
```

```python
d1.display()
```

```python
d1.new_display()
```

```python
d1.display_info()
```

```python
d1.increment()
```

# Method override to extend functionality

```python
class A:
    def __init__(self):
        print("In init of A")
        self.a=90
    def display(self):
        print("This is class A a=", self.a)
    def increment(self):
        self.a += 100
        print("Incremented value is ", self.a)
```

```python
class C(A):
    def __init__(self):
        #call the __init__ of the base class
        # super().__init__()
        A.__init__(self)
        self.c=[9,8,7]
    def display_info(self):
        print(self.c, self.a)
    #over ride increment method to add extra functionality!
    def increment(self):
        self.a += 500
        print("In class C, increment by 500. New val is ", self.a)
```

```python
class E(A):
    def __init__(self):
        #call the __init__ of the base class
        # super().__init__()
        A.__init__(self)
        self.e="extension"
    def display_info(self):
        print(self.e, self.a)
    #over ride increment method to add extra functionality!
    def increment(self):
        A.increment(self) # call base class increment
        #additional functionality
        self.a +=1000
        print("In E class extra 1000 is added", self.a)
```

```python
e=E()
e.increment()
```
