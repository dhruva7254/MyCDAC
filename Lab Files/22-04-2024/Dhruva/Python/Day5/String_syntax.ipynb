# Generate all possible numbers from digits of a given number ( repeatition of digits is allowed )

```python
from itertools  import combinations,permutations
num='1234'
combs=[]
for num_len in range(1,len(num)+1):
    combs += list(set(combinations(num,num_len)))
    # print(list(combs))
perm=[]
for comb in combs:
    perm += list(set(permutations(comb)))
print(list(set(perm)))
```
