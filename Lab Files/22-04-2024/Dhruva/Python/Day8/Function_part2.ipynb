they can be used for dubugging purpose

```python
def odd_num(num):
    if num%2 == 0:
        return False
    else:
        return True
odd_num(5)
```

```python
def odd_num(num,**kwarg):
    """
    function which can give dibugging info
    """
    if ('debug' in kwarg) and (kwarg['debug'] ==1):
        print("Extra info.. num=",num)

    if num%2 == 0:
        if ('debug' in kwarg) and (kwarg['debug'] ==1):
            print("Extra info.. return value",False)
        return False
    else:
        if ('debug' in kwarg) and (kwarg['debug'] ==1):
            print("Extra info.. return value",True)
        return True
odd_num(5,debug=1) #verbose
```

# Logic practice questions


WAP to find unique characters in a sentence and count their occurance.

```python
sent = "IACSD has brilliant students"
sent = sent.lower() # case insensitive
unique_char = set(sent)
print(unique_char)
char_count={}
for char in unique_char:
    char_count[char]  = sent.count(char)
print(char_count)
```

```python
sent = "IACSD has brilliant students"
sent = sent.lower() # case insensitive
char_count={}
for char in sent:
    if char in char_count:
        char_count[char] = char_count[char]+1
    else:
        char_count[char]= 1
print(char_count)
```

# WAP to find unique words in a sentence and count their occurance.

```python
sent = "IACSD has brilliant students trainer is happy with students of iacsd"
sent = sent.lower() # case insensitive
words = sent.split()
word_count={}
for word in words:
    if word in word_count:
        word_count[word] = word_count[word]+1
    else:
        word_count[word]= 1
print(word_count)
```

```python
def check_divisibility(n):
    def inner_function(num):
        if num % n == 0:
            return True
        else:
            return False
    return inner_function

my_fun=check_divisibility(5)
print(my_fun(10),my_fun(11))
my_fun=check_divisibility(7)
print(my_fun(7),my_fun(11))
```

# Decorators

```python
def hello(name,rollno):
    print("Hello .... ", name, rollno)
```

```python
hello("ss",123)
```

```python
import functools

def welcome_decorator(func):
    @functools.wraps(func)
    def wrapper_fun(*arg,**kwarg):
        print("Welcome before executing the function!!!")
        func(*arg,**kwarg)
        print("Thank you for executing the function!")
        return
    return wrapper_fun
```

```python
@welcome_decorator
def hello(name,rollno):
    """This is hello function"""
    print("Hello .... ", name, rollno)
```

```python
hello("ss",123)
```

# non-local scope of variables

there are three types of scope

- global

- local

- non-local

```python
def outer_fun():
    nl_v1=10
    def inner_fun():
        def ii_fun():
            nonlocal nl_v1

            print("inner function")
            print(nl_v1)
            nl_v1+=1
            print(nl_v1)
        ii_fun()
    inner_fun()
outer_fun()
```

## sort a list using lambda function

```python
# simple way of sorting a list
l=[1,5,4,8,2,10,3]
# l.sort()
print(sorted(l))
```

```python
l=[1,5,4,8,2,10,3]

print(sorted(l, key=lambda x: x))# ascending order
print(sorted(l, key=lambda x: -x))# descending order
```

```python
# take list of words and sort using second character of each word

words=["hi","hello","welcome","wonderful"]
print(sorted(words, key = lambda s:s[1]))
```

```python
words=["hi","welcome","hello","wonderful"]
print(sorted(words, key = lambda s:s[1]))
```
